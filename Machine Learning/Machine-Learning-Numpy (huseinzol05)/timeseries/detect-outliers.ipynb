```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
sns.set()
```

```python
tesla = pd.read_csv('TSLA.csv')
tesla.head()
```

```python
signal = np.copy(tesla.Close.values)
std_signal = (signal - np.mean(signal)) / np.std(signal)
```

```python
def detect(signal, treshold = 2.0):
    detected = []
    for i in range(len(signal)):
        if np.abs(signal[i]) > treshold:
            detected.append(i)
    return detected
```

```python
outliers = detect(std_signal)
```

```python
plt.figure(figsize=(15, 7))
plt.plot(np.arange(len(tesla.Close)), tesla.Close)
plt.plot(np.arange(len(tesla.Close)), tesla.Close, 'X', label='outliers',markevery=outliers, c='r')
plt.legend()
plt.show()
```
