```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
N = int(1e4)

X = np.random.uniform(-1, 1, N)
Y = np.random.uniform(-1, 1, N)

inside_radius = X ** 2 + Y ** 2 < 1
pi_estimate = 4 * np.sum(inside_radius) / N
print(pi_estimate)

x_in = X[inside_radius]
y_in = Y[inside_radius]

plt.figure(figsize=(5,5))
plt.scatter(X, Y)
plt.scatter(x_in, y_in)
plt.show()
```
